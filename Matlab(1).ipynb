```python
x= np.linspace(0,5,11)
y= x**2

fig=plt.figure()
plt.plot(x,y,color="red")
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('x1 vs y1')
plt.legend()
plt.show()
```

```python
x= np.linspace(0,5,11)
y= x**2
plt.subplot(3,2,1)
plt.plot(x,y,"r--")
plt.subplot(3,2,6)
plt.plot(y,x,'g*-')
plt.subplot(3,2,5)
plt.plot(x,y,"b--")
```

```python
fig=plt.figure()

axes=fig.add_axes([.1,.1,.8,.8])

axes.plot(x,y,'b')
plt.xlim(1,5)
plt.ylim(1,30)
axes.set_xlabel('set X Label')
axes.set_ylabel('set Y Label')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])# first & 2nd value is for position last two values is size
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x, y, 'b',linewidth=3)
axes1.set_xlabel('x_label_axes2')
axes1.set_ylabel('y_label_axes2')
axes1.set_title('axes 2 title')

axes2.plot(x, y, 'r')
axes2.set_xlabel('x_label_axes2')
axes2.set_ylabel('y_label_axes2')
axes2.set_title('axes 2 title')
```
